## Load Data

```python
df = pd.read_csv('https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/master/glassdoor_jobs.csv')
df.head()
```

## Limpieza de los datos  
---
**Divir columnas**:  
 1. Salary Estimate
 2. Location
 3. Company Name

**Eliminar Columna:**
  - [x] Unnamed:0

**Reemplazar valores:**
 + [x] Founded

```python
#2 location
df['city'] = df.Location.str.split(', ', expand = True)[0]
df.head()
```

```python
df['state'] = df.Location.str.split(', ', expand = True)[1]
df.head()
```

```python
#1 Salary Estimate
df['Salary'] = df['Salary Estimate'].str.split("(", expand= True)[0]
df.head()
```

```python
df['Salary'] = df.Salary.str.replace('K', '')
df.head()
```

```python
df[['salario_min', 'salario_max']] = df.Salary.str.split('-', expand = True)
df.head()
```

```python
df['salario_max'] = df.salario_max.str.replace('1', '0')
df['salario_min'] = df.salario_min.str.replace(' ', '0')
df.head()
```

```python
#Company Name
df['company_name'] = df['Company Name'].str.split('\n', expand=True)[0]
df.head()
```

```python
df.columns
```

```python
#Remplazar Valores de la columna Founded
df['Founded'] = df.Founded.replace(-1, 2020)
```

```python
df.Founded.unique()
```

```python
pd.set_option('display.max_colwidth', 30)
```

```python
df.head()
```

```python
#Eliminar columna Unnamed: 0
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

```python
#Job Description
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r-studio' in x.lower() else 0)
```

```python
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.columns
```

```python
df.to_csv('Salary_data_cleaned.csv', index =False)
```
